```python
from pyspark.sql.functions import col, from_json, expr, row_number, lit, sum, avg, year, month, min, max, count
from pyspark.sql.window import Window
```

```python
SILVER_TRANSACCIONES_TABLE = "silver_transacciones"
GOLD_TRANSACCIONES_TABLE = "gold_transacciones_agg"
```

```python
df_silver = spark.table(SILVER_TRANSACCIONES_TABLE)
display(df_silver)
```

```python
df_gold = (
    df_silver
    .groupBy(
        "canal",
        "operacion",
        year("fecha").alias("a√±o"),
        month("fecha").alias("mes")
        )
    .agg(
        count("*").alias("cantidad"),
        sum("monto").alias("sum_monto"),
        avg("monto").alias("avg_monto"),
        min("monto").alias("monto_min"),
        max("monto").alias("monto_max")
    )
)

display(df_gold)
```

```python
df_gold.write.mode("overwrite").saveAsTable(GOLD_TRANSACCIONES_TABLE)
```
